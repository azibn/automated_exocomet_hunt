```python
def normalise_flux(flux):
    """
    Function: Normalises flux values to 0.
    - Normalisation is x --> (x/mean(x)) - 1

    Parameters:
    :flux (numpy.ndarray): The input flux to be normalized.
    
    Returns:
    :normalised flux (numpy.ndarray): The normalized flux array."""

    flux = np.nan_to_num(flux)
    return flux/flux.mean() - np.ones(len(flux))
```

```python
lc, lc_info = import_XRPlightcurve('/storage/astro2/phrdhx/tesslcs/tesslcs_sector_6_104/2_min_cadence_targets/tesslc_100478849.pkl',sector=6)
```

```python
mean = normalise_flux(lc['corrected flux'])
```

```python
median = lc['corrected flux']/np.median(lc['corrected flux']) - np.ones(len(lc['corrected flux']))
```

```python
plt.figure(figsize=(20,10))
plt.plot(lc['time'],mean,label='mean')
plt.plot(lc['time'],median-0.001,label='median')
plt.legend()
```
