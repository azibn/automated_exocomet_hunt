```python
data, lc_info = import_XRPlightcurve("betapic/xrp/tesslcs_sector_6_104_2_min_cadence_targets_tesslc_270577175.pkl",sector=6,return_type='pandas')
```

```python
count = 0
while count < 1:
    factor = round(random.uniform(0.01,10),1)
    # One choice I make is to have the range of times that I will place my index is +- 1 day from edge of lightcurve/data gap.
    time_range = data['time'][data['time'].between(data['time'].min()+1,data['time'].max()-1, inclusive=False)].reset_index(drop=True) #resets index so consistency is kept when working with indices
    injected_time_index, injected_time = random.choice(list(enumerate(time_range))) # this gives me an index and time value of where I will inject the model (from the first data point of the model, not the dip)
    print(injected_time, "injected time")
    depth = factor * 1.86164653e-03
    y = data['corrected flux']*(1-comet_curve(data['time'],(1.86164653e-03), injected_time, 3.02715600e-01, 3.40346173e-01))
    data['injected_dip_flux'] = data['corrected flux']*(1-comet_curve(data['time'],depth, injected_time, 3.02715600e-01, 3.40346173e-01))
    data['injected_dip_flux_normalised'] = normalise_flux(data['injected_dip_flux']) + 1
    print(factor, "scale factor")
    
    midtransit = data['injected_dip_flux_normalised'][data['time'] == injected_time].values[0]

    if logspace[5] >= midtransit >= logspace[6]:
        print(midtransit, "midtransit")
        count += 1
        break
```

```python
lc_paths =  random.sample(os.listdir("/storage/astro2/phrdhx/tesslcs/tesslcs_sector_6_104/2_min_cadence_targets/"),20)
```

```python
for i in lc_paths:
        data, lc_info = import_XRPlightcurve(
            os.path.join("/storage/astro2/phrdhx/tesslcs/tesslcs_sector_6_104/2_min_cadence_targets/",i), sector=6, return_type="pandas"
        )

        ## scale factor

        factor = round(random.uniform(0.1, 10), 1)


        count = 0
        logspace = np.logspace(0.990,1,10)/10
        midtransit_li = []
        for i in range(len(logspace-1)):
            try:
                while True:
                    ## want the range of transits to be within 1 day from beginning of lightcurve, and one day from the end of the lightcurve
                    time_range = data["time"][
                        data["time"].between(
                            data["time"].min() + 1, data["time"].max() - 1, inclusive=False
                        )
                    ].reset_index(
                        drop=True
                    )  # resets index so consistency is kept when working with indices

                    injected_time_index, injected_time = random.choice(list(enumerate(time_range)))

                    ## multiplying comet depth by scale factor
                    depth = factor * 1.86164653e-03

                    ## comet model (based of Beta Pictoris in Zieba et. al 2019)
                    model = data["corrected flux"] * (
                        1
                        - comet_curve(
                            data["time"], depth, injected_time, 3.02715600e-01, 3.40346173e-01
                        )
                    )
                    data.drop("corrected flux", axis=1, inplace=True)
                    data.insert(
                            1, "corrected flux", model
                    ) 
                    data['normalised flux'] = normalise_flux(data['corrected flux'])
                    midtransit = 1 - data['normalised flux'][data['time'] == injected_time].values[0]
                    if logspace[i] <= midtransit <= logspace[i+1]:
                        midtransit_li.append(midtransit)
                        break
            except:
                pass
        else:
            time_range = data["time"][
                    data["time"].between(
                        data["time"].min() + 1, data["time"].max() - 1, inclusive=False
                    )
                ].reset_index(
                    drop=True
                )  # resets index so consistency is kept when working with indices
            injected_time_index, injected_time = random.choice(list(enumerate(time_range)))

                ## multiplying comet depth by scale factor
            depth = factor * 1.86164653e-03

                ## comet model (based of Beta Pictoris in Zieba et. al 2019)
            model = data["corrected flux"] * (
                    1
                    - comet_curve(
                        data["time"], depth, injected_time, 3.02715600e-01, 3.40346173e-01
                    )
                )
            data.drop("corrected flux", axis=1, inplace=True)
            data.insert(
                        1, "corrected flux", model
                )
```
